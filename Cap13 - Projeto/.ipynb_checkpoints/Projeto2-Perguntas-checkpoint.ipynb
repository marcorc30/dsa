## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
# df_dsa[df_dsa.Categoria == 'Office Supplies'][['Cidade','Valor_Venda']].max()
grupo = df_dsa.groupby('Cidade')['Valor_Venda'].sum()
cidade = grupo.idxmax()
print(f'A cidade com maior valor de venda é {cidade}')
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
# df_dsa.Data_Pedido = df_dsa.Data_Pedido.astype('datetime64')
plt.figure(figsize=(12,6))
vendas_por_data = df_dsa.groupby(['Data_Pedido'])['Valor_Venda'].sum().reset_index().head()

axis = sns.barplot(data = vendas_por_data, x = 'Data_Pedido', y = 'Valor_Venda')
# axis.
# vendas_por_data
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
total_por_estado = df_dsa.groupby('Estado')['Valor_Venda'].sum().reset_index()
total_por_estado.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data = total_por_estado, x = 'Estado', y = 'Valor_Venda').set(title="Vendas por Estado")
plt.xticks(rotation = 80)
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
total_venda_cidade = df_dsa.groupby('Cidade')['Valor_Venda'].sum().sort_values(ascending=False).head(10).reset_index()

ax = sns.barplot(data = total_venda_cidade, x = 'Cidade', y = 'Valor_Venda' ).set(title="Cidades por Venda")
plt.xticks(rotation = 80)
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
def concederDesconto(valorVenda):
    if (valorVenda > 1000):
        return 15
    else:
        return 10
```

```python
df_dsa['Desconto'] = df_dsa['Valor_Venda'].map(concederDesconto)
```

```python
df_dsa[df_dsa.Desconto == 15]['Valor_Venda'].count()
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month 
media_seg_ano_mes = df_dsa.groupby(['Segmento','Ano','Mes'])['Valor_Venda'].mean().reset_index()
media_seg_ano_mes
# media_seg_ano_mes.plot()
sns.lineplot(data=media_seg_ano_mes, x = 'Ano', y = 'Valor_Venda')
# plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python


top_12_subcat = list(df_dsa.groupby('SubCategoria')['Valor_Venda'].sum().sort_values(ascending=False).head(12).reset_index()['SubCategoria']) 
top_12_subcat
total_vendas = df_dsa.query('SubCategoria == @top_12_subcat').groupby(['Categoria', 'SubCategoria'])['Valor_Venda'].sum()
total_vendas.plot()
```
